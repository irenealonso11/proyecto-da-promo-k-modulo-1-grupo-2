#OBJETIVO
Es adivinar la palabra secreta letra por letra antes de que se complete el dibujo del ahorcado.

Â·Reglas del juego:

- SelecciÃ³n de la palabra: El ordenador tiene una lista con una serie de palabras, la palabra elegida se escribe mostrando solo la cantidad de letras de la palabra.

- Adivinanza de letras: El jugador intenta adivinar la palabra diciendo letra por letra 
    - Si la letra esta en la palabra, se revela las posiciÃ³n correspondiente.
    - El jugador puede fallar un maximo de 6 veces.
    - Si la letra no estÃ¡ en la palabra se resta un fallo.

- TerminaciÃ³n del juego: 
    - El juego termina cuando el jugador adivina todas las palabras antes de que se complete el dibujo y gana
    - Si el dibujo se completa antes de que se adivine la palabra, el ordenador gana y se revela la palabra secreta.


```python

# Creamos 2 variables,  fallosðŸ‘Ž y 'aciertos' 

fallos = 6
aciertos = 0

# Creamos la variable "palabra" que serÃ¡ la palabra oculta . 
palabra = "pato"
# Creamos la variable "long_palabra" que nos contarÃ¡ la longitud de la palabra.
# Creamos la variable "palabra_oculta", para que podamos tener la visiÃ³n de la palbra oculta pero representada con asteriscos
# Creamos la variable "lista_aciertos", que nos darÃ¡ una lista con los aciertos, (primero seran todos * y segun vaya acertando se sustituiran por las letras acertadas)
long_palabra= len(palabra)
palabra_oculta= "*" * long_palabra
lista_aciertos = list("*" * long_palabra)
# Primer mensaje que el jugador verÃ¡ (las letras que tiene la palabra, representadas por * cada una)
print(f"La palabra tiene ðŸ”’{palabra_oculta}{long_palabra}letras")

# Realizamos un bucle âž¿ con " while" el cual, continuara mientras la cantidad de fallos estÃ© entre 0 y 6 (variable fallos = 6)
# Pedimos al usuario que indique una letra del abecedario  
# AÃ±adimos en este bucle unas condiciones en funciÃ³n de si la letra estÃ¡ o nÃ³ en la palabra oculta 

while 0 <= fallos <= 6: 
  letra_usuario = input("Indica la letra de la a z")
# AÃ±adimos este print en caso de indicar una letra que ya ha sido acertada
  if letra_usuario in lista_aciertos:
    print("ðŸ¤”Esta letra ya la habÃ­as acertado, Â¡prueba con otra!")
# Utilizamos elif, para definir la 1Âº condicion mientras se acierta la letra.
# Si se cumple esta 1Âº condicion: 
# a) los aciertos aumentan 1 
# b) se lanzarÃ¡ el mensaje Â¡Bien hecho... 
# c) con find() se encontrarÃ¡ la posicion de la letra en la palabra 
# d) El jugador verÃ¡ la posiciÃ³n donde se colocarÃ¡ la letra acertada  
# e) Se actualiza  y se imprime la lista de aciertos  con la letra acertada en la posiciÃ³n correcta  
  elif letra_usuario in palabra:
    aciertos += 1 
    print(f"Â¡Bien hecho! ðŸ‘ La letra '{letra_usuario}' estÃ¡ en la palabra"ðŸŽ‰ðŸŽ‰) 
    posicion = palabra.find(letra_usuario.lower())
    print(f"Posicion letra es {posicion+1}")  
    lista_aciertos[posicion] = letra_usuario
  print(lista_aciertos) 

# Utilizamos if, para definir la 2Âº condicion en la que el jugador no acierta la letra.
# Se restarÃ¡ un fallo 
# Se ha aÃ±adido un if para el conteo de fallos
# Se imprimirÃ¡ el mensaje con las veces que le quedan para fallar (lo siento, la letra ....) 

if letra_usuario not in palabra:
    if fallos >= 1: 
      print(f"Lo siento, ðŸ˜¥ la letra '{letra_usuario}' no estÃ¡ en la palabra.Solo puedes fallar {fallos}veces mÃ¡sðŸ¤”") 
# Utilizamos elif, cuando ha gastado todos los fallos y dara el mensaje: Lo siento, la letra ..... 
    elif fallos == 0:
      print(f"Lo siento, la letra '{letra_usuario}' no estÃ¡ en la palabra.No puedes volver a fallarðŸ¤”")  

# Utilizamos else cuando ya se ha consumido todos los fallos emitiendo el mensaje: Lo siento, no te quedan .... 
#     else: print("Lo siento, no te quedan mÃ¡s oportunidades...Has perdido.") 
# Con la condiciÃ³n if, definimos la igualdad de cuando el nÂº de aciertos es igual a la longitud de la palabra lanzarÃ¡ el mensaje de Felicidades....
# Utilizamos break para romper/parar el bucle ya que el juego ha finalizado. 
if aciertos == long_palabra:
    print("Â¡Felicidades! Has ganado el juego  ðŸ¥‡ðŸ¥‡ðŸ¥‡ ðŸŽ‰ðŸŽ‰ðŸŽ‰")  
    break
```
